```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('/content/ds1.csv')
print('data exported succes')
```

```python
data.head(10)
```

```python
data.tail()
```

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
#prepare the data
X=data.iloc[:, :-1].values
y=data.iloc[:, :1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                            test_size=0.2, random_state=0)
```

Training the Algorithm


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print("Training complete.")
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

Making Predictions

```python
print(X_test) # Testing data - In Hours
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
import pandas as pd

# Assuming y_test and y_pred are arrays or lists
y_test_1d = y_test.flatten() if hasattr(y_test, 'flatten') else y_test
y_pred_1d = y_pred.flatten() if hasattr(y_pred, 'flatten') else y_pred

df = pd.DataFrame({'Actual': y_test_1d, 'Predicted': y_pred_1d})
print(df)
```

```python
df = pd.DataFrame({'Actual': y_test_1d, 'Predicted': y_pred_1d})
print(df)
```

```python
# WE can also test with your own data
hours = 9.30
own_pred = regressor.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

```python
from sklearn import metrics
print('Mean Absolute Error:',
      metrics.mean_absolute_error(y_test, y_pred))
```
